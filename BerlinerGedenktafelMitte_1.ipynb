## Cleaning Step 6: dealing with the values of categorical columns

```python
# Gender:

data['Gender'].value_counts(dropna=False)
```

```python
def cleanGender(x):
    if x in ['m', 'soldiers', 'soldierts','mm','mmmm', 'mmm','mmmmm']:
        return 'm'
    elif x in ['f', 'ff', 'ffffff']:
        return 'f'
    elif x in ['no person']:
        return 'no person'
    elif x in ['several']:
        return 'unspecified'
    else:
        return 'groups of mixed gender'

data['Gender'] = list(map(cleanGender, data['Gender']))
```

```python
data['Gender'].value_counts(dropna=False)
```

```python
# Category:

data['Category'].value_counts(dropna=False)
```

```python
# Activity:

data['Activity'].value_counts(dropna=False)
```

```python
def cleanActivity(x):
    if x in ['Civilian']:
        return 'Civilian'
    elif x in ['Politics', 'Administration']:
        return 'Politics_Administration'
    elif x in ['Writing', 'Liberal_Art', 'Philosophy', 'Singing_Composing_Musician', 'Dancing_Acting_Directing']:
        return 'Creative_Arts'
    elif x in ['Medical_Doctors_Hospital']:
        return 'Medical'
    elif x in ['Public_Place', 'Embassy']:
        return 'PublicPlaces_Buildings'
    elif x in ['Chairman', 'Business_Entrepreneur_Finance']:
        return 'Chairman_Business_Finances'
    elif x in ['Communism']:
        return 'Socialists'
    elif x in ['Military_Police']:
        return 'Military_Police'
    elif x in ['Theologist_Priest', 'Church']:
        return 'Christian_Church'
    elif x in ['Science', 'other_Research', 'Natural_Science']:
        return 'Science_otherResearch'
    elif x in ['Civil Society Organisation']:
        return 'CivilSocietyOrganisation'
    elif x in ['Education_School']:
        return 'Education_School'
    elif x in ['Synagogue_Cementary', 'Rabbi_Judaist']:
        return 'Jewish_Synagogue'
    elif x in ['Architecture', 'Residential', 'Infrastructure_Industrial']:
        return 'Architecture'
    else:
        return 'other'
```

```python
data['Activity'] = list(map(cleanActivity, data['Activity']))
```

```python
data['Activity'].value_counts(dropna=False)
```

```python
# Context:

data['Context'].value_counts(dropna=False)
```

```python
data['Context'] = data['Context'].replace(np.nan, 'no_specific_context')
```

```python
def cleanContext(x):
    if x in ['NS persecuted_victimized', 'NS opponent_resistance']:
        return 'NSresistance_persecuted'
    elif x in ['GDR_victim','GDR_flight']:
        return 'GDRflight_victim'
    elif x in ['Philosopher']:
        return 'Philosopher'
    elif x in ['no_specific_context']:
        return 'no_specific_context'
    elif x in ['Jewish']:
        return 'Jewish'
    elif x in ['Foreign']:
        return 'other_countries'
    elif x in ['royal_prussian', 'Aufklärung_MarchRevolution']:
        return 'Prussian_times'
    elif x in ['War_PoliticalViolence', 'Pacificm']:
        return 'AntiWar'
    elif x in ['NovemberRevolution_Weimar']:
        return 'NovemberRevolution_Weimar'
    elif x in ['Wall_GermanUnity']:
        return 'Wall_GermanUnity'
    elif x in ['social_policy',"worker's rights", 'LGBTQ']:
        return 'SocialPolicy_issues'
    elif x in ['GDR_GDR_personality']:
        return 'GDR_GDR_personality'
    else:
        return 'other'
```

```python
data['Context'] = list(map(cleanContext, data['Context']))
```

```python
data['Context'].value_counts(dropna=False)
```

```python
data.isna().sum()
```

```python
# Material:

data['material'].value_counts(dropna=False)
```

```python
data['material'] = data['material'].replace(np.nan, 'no_information')
```

```python
def cleanMaterial(x):
    if x in ['Edelmetall', 'Metall', 'Metallguss', 'Edelstahl', 'Stahl']:
        return 'Metall_Stahl'
    elif x in ['Stein', 'Grabstein', 'Granit', 'Muschelkalkstein', 'Schiefer', 'Mauer', 'Marmor']:
        return 'Stein'
    elif x in ['Infokasten', 'Informationstafel']:
        return 'Infokasten'
    elif x in ['Bronze', 'Messing', 'Kupfer']:
        return 'Bronze, Messing, Kupfer'
    elif x in ['KPM', 'Porzellan']:
        return 'Porzellan'
    elif x in ['Acrylglas']:
        return 'Acrylglas'
    elif x in ['Mauerkreuz']:
        return 'Mauerkreuz'
    else:
        return 'other'

data['material'] = list(map(cleanMaterial, data['material']))
```

```python
data['material'].value_counts(dropna=False)
```

```python
def cleanBirthplace(x):
    if 'Berlin' in x:
        return 'Berlin'
    elif 'Leipzig' in x:
        return 'Leipzig'
    elif 'Potsdam' in x:
        return 'Potsdam'
    elif 'Breslau' in x:
        return 'Breslau'
    elif 'no_information' in x:
        return 'no_information'
    else:
        return 'other'
```

```python
data['birthplace'] = list(map(cleanBirthplace, data['birthplace']))
```

```python
Berlinlist = [ 'Berlin', 'Wannsee', 'Steglitz', 'Schöneberg', 'Henningsdorf']

for x in Berlinlist:
    data.loc[data['deathplace'].str.contains(x), 'deathplace'] = 'Berlin'
```

```python
NScapture=['KZ', 'Zuchthaus', 'Plötzensee', 'Theresienstadt', 'Auschwitz', 'Ukraine']

for x in NScapture:
    data.loc[data['deathplace'].str.contains(x), 'deathplace'] = 'KZ or NS_capture'
```

```python
Foreign=['Paris', 'Schweiz', 'Schweden', 'Jerusalem', 'Kopenhagen', 'Stockholm', 'Genf', 'Russland', 'Manila', 'Gstaad', 'Connecticut', 'Wien', 'New York', 'New Jersey', 'Sankt Petersburg', 'Grönland', 'Zürich', 'Frankreich', 'Cambridge', 'Memphis', 'London']

for x in Foreign:
    data.loc[data['deathplace'].str.contains(x), 'deathplace'] = 'outside Germany'
```

```python
def cleanDeathplace(x):
    if 'Berlin' in x:
        return 'Berlin'
    elif 'KZ or NS_capture' in x:
        return 'KZ or NS_capture'
    elif 'outside Germany' in x:
        return 'outside Germany'
    elif 'Baden-Baden' in x:
        return 'Baden-Baden'
    elif 'no_information' in x:
        return 'no_information'
    else:
        return 'other in Germany'
```

```python
data['deathplace'] = list(map(cleanDeathplace, data['deathplace']))
```
